```python
import timm
import tqdm as notebook_tqdm
model = timm.create_model('darknet53', pretrained=True)
```

```python
import torch
from torchvision import models
import torch.utils.model_zoo as model_zoo

model_paths = {
    'darknet19': 'https://s3.ap-northeast-2.amazonaws.com/deepbaksuvision/darknet19-deepBakSu-e1b3ec1e.pth'
}

model=torch.load_state_dict(model_zoo.load_url(model_paths['darknet19'],  progress=True))

print('Model is loaded')
```

```python
print(model.feature_info)
```

```python
import torch

x = torch.randn(1, 3 ,320, 160)  # batch of images
features = model.forward(x)

# Suppose you pick the last 3 stages:
feature_small = features[2]   # medium resolution
feature_medium = features[3]  # lower resolution
feature_large = features[4]   # lowest resolution, highest-level features

print([f.shape for f in [feature_small, feature_medium, feature_large]])
#[torch.Size([1, 128, 80, 40]), torch.Size([1, 256, 40, 20]), torch.Size([1, 512, 20, 10])]
```

```python
import torch
from pathlib import Path

save_path = Path("darknet53_feat.pth")
save_path.parent.mkdir(parents=True, exist_ok=True)

# 3️⃣ Save the full model
torch.save(model.state_dict(), save_path)
print(f"Model saved locally to {save_path}")
```
